```python
import pandas as pd
import numpy as np

# 날짜형 변환
df['start_time_di'] = pd.to_datetime(df['start_time_di'], errors='coerce')
df['last_event_di'] = pd.to_datetime(df['last_event_di'], errors='coerce')

# [파생] 전체기간(시작일 포함)
df['total_period_days'] = (df['last_event_di'] - df['start_time_di']).dt.days + 1

# 혹시 이상치로 0 이하가 나오면 결측 처리
df.loc[df['total_period_days'] <= 0, 'total_period_days'] = np.nan

# [파생] 전체기간 중 실제 활동 비율
df['active_day_ratio'] = df['ndays_act'] / df['total_period_days']
```
